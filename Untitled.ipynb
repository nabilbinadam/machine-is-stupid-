```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  # Corrected the typo here
import matplotlib.pyplot as plt  # Corrected the import statement here
```

```python
# lets create data and arrange
X= np.arange(10).reshape((5,2))
X
```

```python
y= range(5)
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train
```

```python
model=LinearRegression()
model
```

```python

#fit() is a function to train 
model.fit(X_train,y_train)
```

```python
# predict() is to predict 
prediction=model.predict(X_test)
prediction
```
